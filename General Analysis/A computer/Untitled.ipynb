```python
import numpy as np
import matplotlib.pyplot as plt

# Define f(x) as a piecewise function
def f(x):
    if x > 0:
        return np.exp(-1 / x)
    else:
        return 0

# Define g(x)
def g(x):
    return f(x) / (f(x) + f(1 - x))

# Define h(x) = g(1 + x) * g(1 - x)
def h(x):
    return g(2 + x) * g(2 - x)

# Generate x values from -2 to 2
x_values = np.linspace(-2, 2, 400)
h_values = [h(x) for x in x_values]

# Plot the function
plt.figure(figsize=[8, 4])
plt.plot(x_values, h_values, label='h(x) = g(1 + x) * g(1 - x)')
plt.title('Plot of h(x)')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the bump function h(x) using vectorized operations
def f(x):
    return np.where(x > 0, np.exp(-1 / x), 0)

def g(x):
    return f(x) / (f(x) + f(1 - x))

def h(x):
    return g(2 + x) * g(2 - x)

# 2D version of H(x, y) = h(sqrt(x^2 + y^2))
def H(x, y):
    r = np.sqrt(x**2 + y**2)  # Radial distance
    return h(r)

# Create a grid of x, y values
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# Compute H(X, Y)
Z = H(X, Y)

# Plot the 2D bump function
plt.figure(figsize=[8, 6])
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='H(x, y)')
plt.title('2D Plot of H(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
